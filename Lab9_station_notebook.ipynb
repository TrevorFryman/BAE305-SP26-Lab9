# Lab 9 — Station list and map
Run the notebook cells in order. Each cell performs one step: load data, list station names, import helper functions and filter, extract unique locations, generate/save the interactive map, then embed the saved map and provide a link.

```python
# 1) Imports and load CSV
import pandas as pd
from IPython.display import display, HTML, IFrame

# Load the station CSV (run this cell first)
df = pd.read_csv('station.csv', low_memory=False)
print('Loaded station.csv — rows:', len(df))
```

```python
# 2) List unique station names
if 'MonitoringLocationName' in df.columns:
    stations = df['MonitoringLocationName'].dropna().unique()
    print(f'Found {len(stations)} unique station names:')
    for s in stations:
        print('-', s)
else:
    print('Column "MonitoringLocationName" not found in station.csv')
```

```python
# 3) Import helper functions and filter for water-quality stations
import lab9_station_map as l9

filtered = l9.filter_water_quality(df)
print('Filtered rows:', len(filtered))
display(filtered.head())
```

```python
# 4) Extract unique station locations
locations = l9.get_station_locations(filtered)
print('Unique locations:', len(locations))
display(locations.head())
# Optional: save unique locations to CSV
locations.to_csv('unique_locations.csv', index=False)
print('Saved unique_locations.csv')
```

```python
# 5) Generate and save the interactive folium map
m = l9.plot_and_show_map(locations, out_html='station_map.html', open_in_browser=False)
print('Map saved to station_map.html')
```

```python
# 6) Embed the saved interactive map and provide link
display(IFrame('station_map.html', width=900, height=600))
print('Interactive map file: station_map.html')
print('Open this file in your browser for the full interactive experience.')
```
